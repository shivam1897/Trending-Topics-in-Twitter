```python
#importing Libraries
import  pandas as pd
import re
import emoji
import nltk
```

```python
#downloading dependancies
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
```

```python
def give_emoji_free_text( text):
    allchars = [str for str in text]
    emoji_list = [c for c in allchars if c in emoji.UNICODE_EMOJI]
    clean_text = ' '.join([str for str in text.split() if not any(i in str for i in emoji_list)])

    return clean_text

def remove_emoji(df):
    for col in range(len(df)):
        temp = df['text'][col]
        df['text'][col]= give_emoji_free_text(temp)
        
def remove_url(df):
    for col in range(len(df)):
        df['text'][col] = re.sub(r'http\S+', '', df['text'][col])
        

def remove_stop_words(df):
    temp = [];
    new = []

    for col in range(len(df)):
        temp = df['text'][col].split()
        for j in temp:
            if(j not in stop):
                #print(j)
                new.append(j)
        df['text'][col] = ' '.join(new)
        temp = []
        new = []
```

```python
#reading csv file

filepath = 'new.csv'
data = pd.read_csv(filepath)

#extracting text column and converting it to a dataframe
old = data['text'].copy()
df = pd.Series.to_frame(old)

remove_url(df)
remove_emoji(df)

#removing stop words
stop = stopwords.words('english')
remove_stop_words(df)

#removing pattern such as 'RT @blablabla:'

dumy = df.copy()
for col in range(len(df)):
    dumy['text'][col] = re.sub(r'@\w*?:|RT', '', dumy['text'][col])
    print(dumy['text'][col])


```
